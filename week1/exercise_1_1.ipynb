- Split data into a train and test. Use 60 percent data in the training and test set which is assigned
  - randomly
  - assigned by the first 60 percent as train and the rest as test. 

```python
from sklearn.model_selection import train_test_split
# So we want to split the data into a training and a test set using 2 different methods

# First let's split the data randomly using a 60% training and 40% test split
df_X_train_random, df_X_test_random, df_y_train_random, df_y_test_random = train_test_split(df_X, df_y, test_size=0.4, random_state=1)

# Now let's split the data naively by taking the first 60% of the data as the training set and the last 40% as the test set
train_rows = int(len(df_X) * 0.6)
df_X_train_naive = df_X[:train_rows]
df_X_test_naive = df_X[train_rows:]
df_y_train_naive = df_y[:train_rows]
df_y_test_naive = df_y[train_rows:]
```
